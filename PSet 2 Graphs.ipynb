```python
# imports

import matplotlib.pyplot as plt
import numpy as np

# constants
Z = 1
a_0 = 0.5291
```

```python
# Equations

def R(n, l, r):
    if l == 0:
        if n == 1:
            return 2 * ((Z / a_0) ** 1.5) * np.e ** (- (Z * r) / a_0)
        if n == 2:
            return 2 * ((Z / (2 * a_0)) ** 1.5) * (1 - (Z * r) / (2 * a_0)) * np.e ** (- (Z * r) / (2 * a_0))
        if n == 3:
            return 2 * ((Z / (3 * a_0)) ** 1.5) * (1 - (2 * Z * r) / (3 * a_0) + (2 * Z * Z * r * r) / (27 * a_0 * a_0)) * np.e ** (- (Z * r) / (3 * a_0))

    if l == 1:
        if n == 2:
            return (1 / np.sqrt(3)) * ((Z / (2 * a_0)) ** 1.5) * ((Z * r) / a_0) * np.e ** (- (Z * r) / (2 * a_0))
        if n == 3:
            return (4 * np.sqrt(2) / 3) * ((Z / (3 * a_0)) ** 1.5) * ((Z * r) / a_0) * (1 - (Z * r) / (6 * a_0)) * np.e ** (- (Z * r) / (3 * a_0))
    
    if l == 2:
        if n == 3:
            return ((2 * np.sqrt(2)) / 27 * np.sqrt(5)) * ((Z / (3 * a_0)) ** 1.5) * ((Z * r) / a_0) ** 2 * np.e ** (- (Z * r) / (3 * a_0))
    
def Y(l, m, r, x, y, z):
    if l == 0 and m == 0:
        return np.sqrt(1/(4 * np.pi))
    if l == 1 and m == 0:
        return np.sqrt(3 / (4 * np.pi)) * z / r
    if l == 2 and m == 0:
        return np.sqrt(5 / (16 * np.pi)) * (2 * z ** 2 - x ** 2 - y ** 2) / r ** 2
    
def wave_func(n, l, m, r, x, y, z):
    return R(n, l, r) * Y(l, m, r, x, y, z)
```

```python
r = np.linspace(0, 5, num=10000)
plt.plot(r, 2 * wave_func(1, 0, 0, r, 0, 0, 0), label="1s")
plt.plot(r, 2 * wave_func(2, 0, 0, r, 0, 0, 0), label="2s")
plt.plot(r, 2 * wave_func(3, 0, 0, r, 0, 0, 0), label="3s")
plt.title("Problem 1(a), Probability Amplitude")
plt.legend()
plt.grid()
plt.show()
```

```python
plt.plot(r, (wave_func(1, 0, 0, r, 0, 0, 0)) ** 2, label="1s")
plt.plot(r, (wave_func(2, 0, 0, r, 0, 0, 0)) ** 2, label="2s")
plt.plot(r, (wave_func(3, 0, 0, r, 0, 0, 0)) ** 2, label="3s")
plt.legend()
plt.grid()
plt.title("Problem 1(a), Probability Density")
plt.show()
```

```python
r = np.linspace(0.001, 15, num=100)
plt.plot(r, (r ** 2) * np.abs(wave_func(1, 0, 0, r, 0, 0, 0)) ** 2, label="1s")
plt.plot(r, (r ** 2) * np.abs(wave_func(2, 0, 0, r, 0, 0, 0)) ** 2, label="2s")
plt.plot(r, (r ** 2) * np.abs(wave_func(3, 0, 0, r, 0, 0, 0)) ** 2, label="3s")
plt.legend()
plt.grid()
plt.title("Problem 1(b), Radial Probability Density")
plt.show()
```

```python
plt.plot(r, (r ** 2) * np.abs(wave_func(2, 0, 0, r, 0, 0, 0)) ** 2, label="2s")
plt.plot(r, (r ** 2) * np.abs(wave_func(2, 1, 0, r, 0, 0, r)) ** 2, label="$2p_z$ along z-axis")
plt.legend()
plt.grid()
plt.title("Problem 1(c), Radial Probability Densities")
plt.show()
```

```python
plt.plot(r, (r ** 2) * np.abs(wave_func(2, 1, 0, r, r, 0, 0)) ** 2, label="$2p_z$ along x-axis")
plt.plot(r, (r ** 2) * np.abs(wave_func(2, 1, 0, r, 0, 0, r)) ** 2, label="$2p_z$ along z-axis")
plt.legend()
plt.grid()
plt.title("Problem 1(d), Radial Probability Density")
plt.show()
```

```python
plt.plot(r, (r ** 2) * np.abs(wave_func(3, 0, 0, r, 0, 0, r)) ** 2, label="3s")
plt.legend()
plt.grid()
plt.title("Problem 1(e), 3s Close Up")
plt.show()

plt.plot(r, (r ** 2) * np.abs(wave_func(3, 0, 0, r, 0, 0, r)) ** 2, label="3s")
plt.plot(r, (r ** 2) * np.abs(wave_func(3, 1, 0, r, 0, 0, r)) ** 2, label="$3p_z$")
plt.plot(r, (r ** 2) * np.abs(wave_func(3, 2, 0, r, 0, 0, r)) ** 2, label="$3d_{z^2}$")
plt.legend()
plt.grid()
plt.title("Problem 1(e), Radial Probability Density resolved along z-axis")
plt.show()
```
